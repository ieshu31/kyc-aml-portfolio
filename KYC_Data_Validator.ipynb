```python
import pandas as pd

df=pd.read_csv('KYC_Case_Tracker.csv')
print("Total customers:", len(df))
print(df.head())

sanctioned=["Iran","North Korea","Cuba","Syria","Venezuela","Sudan"]
print("\n--sanctioned country flags--")
flagged=df[df['Country'].isin(sanctioned)]
print(flagged[['Customer Name','Country','Risk Rating']])
print("Total Flagged:", len(flagged))

print("\n--- MISSING DOCUMENTS ---")
df['Documents Received'] = df['Documents Received'].str.strip()
missing_docs = df[df['Documents Received'] == 'No']
print(missing_docs[['Customer Name', 'Country', 'Documents Received']])
print("Total missing:", len(missing_docs))

print("\n--- HIGH RISK CUSTOMERS ---")
df['Risk Rating'] = df['Risk Rating'].str.strip()
high_risk = df[df['Risk Rating'] == 'High']
print(high_risk[['Customer Name', 'Country', 'Risk Rating']])
print("Total high risk:", len(high_risk))

print("\n--- GENERATING EXCEPTION REPORT ---")

# Combine all flags into one report
exceptions = df[
    (df['Country'].isin(sanctioned)) |
    (df['Documents Received'] == 'No') |
    (df['Risk Rating'] == 'High')
]

print(exceptions[['Customer Name', 'Country', 'Risk Rating', 'Documents Received']])
print("\nTotal exceptions flagged:", len(exceptions))

# Export to CSV
exceptions.to_csv('exception_report.csv', index=False)
print("Exception report saved!")

```
